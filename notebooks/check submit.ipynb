```python
import sys
import glob
from os.path import splitext, basename
from datetime import datetime

import numpy as np
import pandas as pd

%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
```

```python
def random_image(folder='index'):
    path = '/media/hdd1/kaggle/landmark-retrieval-challenge/{}/{}.jpg'
    if folder == 'index':
        filename = np.random.choice(index_images)
    elif folder == 'test':
        filename = np.random.choice(query_images)        
    print(filename)
    plt.imshow(mpimg.imread(path.format(folder, filename)))
    plt.show()
    
def plot_image(filename, folder='test'):
    path = '/media/hdd1/kaggle/landmark-retrieval-challenge/{}/{}.jpg'
    plt.imshow(mpimg.imread(path.format(folder, filename)))
    plt.show()
    
def plot4images(filenames):
    path = '/media/hdd1/kaggle/landmark-retrieval-challenge/index/{}.jpg'
    filenames = filenames.split(' ')
    plt.figure(1)
    for i, x in enumerate([221, 222, 223, 224]):
        plt.subplot(x)
        plt.imshow(mpimg.imread(path.format(filenames[i])))
    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25, wspace=0.35)
    plt.show()
    
def compare(query_id, filenames):
    path_query = '/media/hdd1/kaggle/landmark-retrieval-challenge/test_rescale/{}.jpg'
    path_index = '/media/hdd1/kaggle/landmark-retrieval-challenge/index_rescale/{}.jpg'
    filenames = filenames.split(' ')
    plt.figure(1, figsize=(10, 10))
    plt.subplot(221)
    plt.imshow(mpimg.imread(path_query.format(query_id)))
    plt.axis('off')
    for i, x in enumerate([222, 223, 224]):
        plt.subplot(x)
        plt.imshow(mpimg.imread(path_index.format(filenames[i])))
        plt.axis('off')
    plt.show()
```
