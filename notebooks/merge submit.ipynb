```python
from datetime import datetime
import pandas as pd
import numpy as np
```

```python
submit1 = pd.read_csv('../submit/16_dir_desc/submit_2018-04-17_15.13.32_dir_full_index_qe_2_0.424.csv.gz')
submit2 = pd.read_csv('../submit/10_flatindex_delf_vlad/ssr_k256/submit_2018-04-06_18.43.46_vlad_full_index.csv.gz')
```

```python
final = {}
for img_id, preds1, preds2 in zip(submit1['id'].values, submit1['images'].values, submit2['images'].values):
    if not isinstance(preds1, str):
        continue
    preds = []
    preds1 = preds1.split(' ')
    preds2 = preds2.split(' ')
    i = 0
    while len(preds) < 100:
        preds.append(preds1[i])
        if preds1[i] != preds2[i]:
            preds.append(preds2[i])
        i += 1
    final[img_id] = ' '.join(preds)
```

```python
submit = pd.read_csv('../data/sample_submission.csv')
submit['images'] = submit['id'].map(final)
# submit['images'].apply(lambda x: len(x.split(' ')) if isinstance(x, str) else 0)
```

```python
submit_filename = '../submit/merge/submit_{}_merge_10_ssr256_16_dir_v2.csv.gz'.format(
    datetime.now().strftime('%Y-%m-%d_%H.%M.%S'))
submit.to_csv(submit_filename, index=False, compression='gzip')
```

```python
submit1['images2'] = submit2['images'].values
def process(x):
    preds1, preds2 = x
    if not isinstance(preds1, str):
        return np.nan
    preds1 = preds1.split(' ')[5:]
    preds2 = preds2.split(' ')[:5]
    return '{} {}'.format(' '.join(preds1), ' '.join(preds2))
    
submit1['images'] = submit1[['images', 'images2']].apply(process, axis=1)
submit1.drop('images2', 1, inplace=True)
```

```python
predicted = pickle_load('../submit/17_dir_dbe/predicted.pkl')
train_filenames = pickle_load('../submit/17_dir_dbe/filenames_index.dump')
test_filenames = pickle_load('../submit/17_dir_dbe/test_filenames.pkl')
```

```python
import pickle
def pickle_load(path):
  """function to load pickle object"""
  with open(path, 'rb') as f:
    return pickle.load(f, encoding='latin1')
```

```python
def _make_results(predicted):
    results = {}
    for query_ix, query_result in enumerate(predicted):
        results[test_filenames[query_ix]] = ' '.join(
        [train_filenames[ix] for ix in query_result if ix != -1])
    return results
```

```python
res = _make_results(predicted)
```

```python
submit = pd.read_csv('../data/sample_submission.csv')
submit['images'] = submit['id'].apply(lambda ix: res.get(ix, ''))
submit_filename = '../submit/17_dir_dbe/submit_{}_dir_dbe_100.csv.gz'.format(datetime.now().strftime('%Y-%m-%d_%H.%M.%S'))
submit.to_csv(submit_filename, index=False, compression='gzip')
```
